```python
# Shuffle the array
def shuffle(nums, n):
    out_lst = []
    for i in range(n):
        out_lst.extend([nums[i],nums[n+i]])
    return out_lst

shuffle([1,4,5,6,7,8],3)
```

```python
# Running Sum of 1D array
def runningSum(nums):
    out_nums = []
    temp = 0
    for i in nums:
        temp+=i
        out_nums.append(temp)
    return out_nums

runningSum([1,3,5,7,9])
```

```python
# Kids with the greatest number of candies
def kidsWithCandies(candies, extraCandies) :
    outlst = []
    for i in candies:
        if i+extraCandies >= max(candies):
            outlst.append(True)
        else:
            outlst.append(False)
    return outlst

candies = [4,2,1,1,2]
extraCandies = 2
kidsWithCandies(candies,extraCandies)
```
