```python
# Find numbers with even number of digits
nums = [555,901,482,1771,111111]
print(sum(1 for i in nums if len(str(i)) % 2 == 0))
```

```python
# Number of good pairs.
nums = [1,2,3,1,1,3]
count = 0
for i in range(len(nums)):
    for j in range(i+1,len(nums)):
        if nums[i] == nums[j]:
            count+=1
count
```

```python
# How many numbers in the array are smaller than the current number
nums = [8,1,2,2,3]
lst = []
for i in nums:
    count = 0
    for j in nums:
        if i > j:
            count+=1
    lst.append(count)
lst
```

```python
# Subtract the product of the digits and the sum of its digits   
n = 234  
product = 1
total = 0

while n > 0:
    digit = n % 10
    product *= digit
    total += digit
    n //= 10
print(product - total)
```

```python
# XOR operation in an array
n = 5
start = 0
lst = list(start+2*i for i in range(n))
res = lst[0]
for i in lst[1:]:
    res = res^i
res
```
