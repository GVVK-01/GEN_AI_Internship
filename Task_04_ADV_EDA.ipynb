```python
# Maximum product of two elements in an array
nums = list(map(int, input().split()))
nums.sort()
print((nums[-1]-1)*(nums[-2]-1))
```

```python
# Count number of teams
rating = list(map(int, input().split()))
count = 0
for i,val in enumerate(rating[1:-1]):
    llv = rgv = lgv = rlv = 0
    for j in rating[:i+1]:
        if j < val :
            llv+=1
        elif j > val :
            lgv+=1
    for k in rating[i+1:]:
        if k > val:
            rgv += 1
        elif k < val:
            rlv += 1
    count += llv*rgv + lgv*rlv
print(count)
```

```python
# Number of students doing homework at a given time
startTime = list(map(int, input().split()))
queryTime = int(input())
endTime = list(map(int, input().split()))
count = 0
for i in range(len(startTime)):
    if startTime[i] <= queryTime <= endTime[i]:
        count+=1
count
```

```python
# Number of steps to reduce a number to zero
num = int(input())
count = 0
while num !=0:
    if num%2 == 0:
        num = num/2
        count +=1
    else:
        num = num - 1
        count+=1
count
```

```python

# Counting bits
num = int(input())
lst = []
for i in range(num+1):
    lst.append(bin(i)[2:].count('1'))
lst
```
